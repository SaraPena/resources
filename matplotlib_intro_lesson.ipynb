```python
from random import randint
```

```python
#Some random data
x1 = [randint(-5,25) + n for n in range(150)]
x2 = [x + randint(-6,6) for x in x1]
```

```python
# we can use the alpha kwarg to define how transparent the color is
plt.plot(x1, c = 'green', alpha = 0.6)
plt.plot(x2, c = 'red', alpha = 0.4)
plt.show()
```

```python
# Plotting our random data from the last step
plt.scatter(range(len(x1)), x1)
plt.scatter(range(len(x2)), x2)
plt.show()
```

```python
# We can change the size of the points with the s keyword argument, and like in previous examples, we can chage the color with c.
plt.scatter(range(len(x1)), x1, s=30, c = 'red')
plt.scatter(range(len(x2)), x2, s=10, c ='green')
plt.show()
```

```python
#Titles can be added with the .title method, and axis labels with the .xlabel and .ylabel methods.
plt.scatter(range(len(x1)), x1, s=30, c = 'red')
plt.scatter(range(len(x2)), x2, s=10, c = 'green')
plt.title('A couple of random series')
plt.show()
```

```python
x = list(range(-49,50))
y = [n**2 for n in x]
plt.scatter(x,y)
plt.title('A Quadratic Distribition, $y = x^2$')
plt.xlabel('$x$')
plt.ylabel('$x^2$')
plt.show()
```

```python
x = [1, 2, 3.14, 4]
y = [n**2 for n in x]
plt.scatter(x,y)
plt.xticks([1, 2, 3.14, 4], ['one', 'two','$\pi$', 'four'])
plt.show()
```

```python
#We can change the type of the line on a line chart with the ls (line style) keywork argument.
import math
x = range(-4,4)
y = [math.sin(n) for n in x]
plt.scatter(x,y)
plt.plot(x, y, ls = ':')
plt.title('An approximation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
#Axis ticks can be rotated with the rotation argument.
x = range(1,5)
y = x
plt.scatter(x,y)
plt.yticks([1, 2, 3, 4],['one', 'two', 'three', 'four'], rotation = 45)
plt.xticks(rotation = 45)
plt.show()
```

```python
import random
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] *2
x = range(14)
y = [random.randint(1,10) for _ in x]
plt.plot(x,y)
plt.xticks(x, weekdays, rotation = 60)
plt.show()
```

```python
#add text labels

x = list(range(-4, 4))
x.append(3.14159)
y = [math.sin(n) for n in x]
plt.scatter(x, y)
plt.plot(x, y, ls=':')
plt.text(0.25, 0, ('0,0'), fontsize = 14, color = 'blue')
plt.text(1.2, 0.1, '($\pi$, sin($\pi$))', fontsize=14, color = 'firebrick')
plt.show()
```

```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c = 'orange')
plt.plot(x, z, c = 'green')

plt.annotate('Intersection', xy = (0,0), xytext = (-3,5), arrowprops = {'facecolor':'blue'})
plt.show()
```

```python
#Figure Size and Legends

x = range(-4 ,5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize = (14, 6)) #(width, height)

plt.scatter(x, sin_x, c = 'blue')
plt.scatter(x, cos_x, c = 'orange')
plt.plot(x, sin_x, c = 'blue', label = '$\sin(x)$')
plt.plot(x, cos_x, c = 'orange', label ='$\cos(x)$')

plt.legend(loc = 'upper right')
plt.title('Approximate sine and cose curves')
plt.xlabel('$x$')

plt.show()
```

```python
# Using subplots is very common, and the number of rows, columns,and and active chart can be specified in a short hand of just one number.

# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(121)
plt.plot(x, y)
plt.title('x ~ z')

# the second subplot
plt.subplot(122)
plt.plot(x, z)
plt.title('x ~ z')

plt.show()
```

```python
# Histograms

#We can create histograms with the .hist function.

x = [1, 1, 2, 3, 3, 3, 4, 4, 5]
plt.hist(x)
plt.show()
```

```python
# we can also explicitly specify bins:

x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x, bins =[0, 1, 2, 3, 4, 5, 6])
plt.show()
```

```python
# Alignments: left, right, middle

x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt. figure(figsize=(16,6))
plt.suptitle(f'x = {x}')

plt.subplot(131)
plt.hist(x, bins =[0, 1, 2, 3, 4, 5, 6])
plt.title('align = "middle" (default)')

plt.subplot(132)
plt.hist(x, bins = [0, 1, 2, 3, 4, 5, 6], align = 'left')
plt.title('align = "left"')

plt.subplot(133)
plt.hist(x, bins = [0, 1, 2, 3, 4, 5, 6], align = 'right')
plt.title('align = "right"')

plt.show()
```

```python
# We can plot multiple histograms together.
import random

x1 = [randint(1,5) for _ in range(20)]
x2 = [randint(1,5) for _ in range(29)]

plt.hist(x1, bins = [0, 1, 2, 3, 4, 5, 6], align = 'left', edgecolor = 'black', alpha = 0.5, color = 'blue')
plt.hist(x2, bins = [0, 1, 2, 3, 4, 5, 6], align = 'left', edgecolor = 'black', alpha = 0.5, color = 'blue')

plt.show()
```

```python
# with a legend.

x1 = [randint(1,5) for _ in range(20)]
x2 = [randint(1,5) for _ in range(20)]

plt.hist(x1, bins = [1, 2, 3, 4, 5, 6], align = 'left', edgecolor = 'black', alpha = 0.4, label = '$x_1$')
plt.hist(x2, bins = [1, 2, 3, 4, 5 ,6], align = 'left', edgecolor = 'black', alpha = 0.4, label = '$x_2$')
plt.legend()


plt.show()
```

```python
# we can also plot the series series side by side by passing the data series as a tuple to the hist function.
x1 = [randint(1,5) for _ in range(20)]
x2 = [randint(1,5) for _ in range(20)]

plt.hist((x1, x2), bins = [1, 2, 3, 4, 5, 6], align = 'left', edgecolor = 'black', alpha = 0.6, label = ['$x_1$', '$x_2$'])

plt.legend()
plt.show()
```

```python
# histogram subplots.

plt.figure(figsize = (12,8))
for i in range(1,5):
    n = 10**i
    x1 = [randint(1, 6) for _ in range(n)]
    x2 = [randint(1, 6) for _ in range(n)]
    
    total = [x + y for x, y in zip(x1,x2)]
    
    plt.subplot(2,2,i)
    plt.hist(total, bins = range(2,14), align = 'left', color = 'white', edgecolor = 'black')
    plt.title(f'n = {n}')
    
plt.suptitle('Sum of 2 dice rolls')
plt.show()
```

```python
# We can also create 2d histograms

n = 1000
x1 = [randint(1, 6) for _ in range(n)]
x2 = [randint(1, 6) for _ in range(n)]

plt.hist2d(x = x1, y = x2, bins = range(1,8), )
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
## Axis Scales
# We can change the scales of the axes to be, for example, logarithmic.

x = [n / 100 for n in range(-10000,10000)]
y = list(x)

plt.plot(x,y)
plt.yscale('log')
plt.show()
```
